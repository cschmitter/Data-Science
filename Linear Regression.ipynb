### How to define delta operator...?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def h_function(X, theta, y):
    return 0.5 * ((X @ theta - y)**2)

def delta_h(X, theta, y):
    X_t = np.transpose(X)
    return 2/np.size(X.T[0]) * (X.T @ (X @ theta - y))
```

```python
rng = np.random.default_rng(10300)
m = 10000
error = 0.0001
alpha = 0.01
```

```python
a = rng.uniform(size=(m,1))
b = rng.uniform(size=(m,1))
noise = rng.uniform(size=(m,1), high=0.5, low=-0.5)
y = 3 * a - 2 * b + noise + 10
X = np.concatenate((a, b, np.ones((m,1))), axis = 1)
#print(X)
#print(y)
```

```python
theta_0 = rng.random((3,1))
#print(theta_0)
```

```python
def predict(theta, X):
    return X @ theta
```

## Test of functions

```python
def train(X, theta, y, n):
    theta = theta_0[:]
    for i in range(n):
        theta = theta - alpha * delta_h(X, theta, y)
    return theta
```

```python
theta = train(X, theta_0, y, 1000)
plt.scatter(x=X.T[0], y=X.T[1], c=2*y)
```
